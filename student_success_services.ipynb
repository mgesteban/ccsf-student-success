## Imports & Setup

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 50)
```

## Raw files from the Data Mart

```python
file1 = "2014_2016_student_success_services.csv"
file2 = "2017_2020_student_success_services.csv"
file3 = "2021_2025_student_success_services.csv"

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
```

## Inspect the First Few Rows

```python
df1.head(10)
```

```python
df2.head(10)
df3.head(10)
```

## Drop Metadata Rows

```python
df1_clean = df1.iloc[3:].reset_index(drop=True)
df2_clean = df2.iloc[3:].reset_index(drop=True)
df3_clean = df3.iloc[3:].reset_index(drop=True)
```

```python
df1_clean.head() #for inspection
```

## Rename the Service Column

```python
df1_clean.rename(columns={df1_clean.columns[1]: "service"}, inplace=True)
df2_clean.rename(columns={df2_clean.columns[1]: "service"}, inplace=True)
df3_clean.rename(columns={df3_clean.columns[1]: "service"}, inplace=True)
```

```python
df1_clean["service"].head()
```

## Clean and Simplified Service Names

```python
def clean_service_name(s):
    if pd.isna(s):
        return s
    s = s.lower()
    if "counsel" in s:
        return "counseling"
    if "education plan" in s or "ed plan" in s:
        return "ed_plan"
    if "orientation" in s:
        return "orientation"
    if "assessment" in s:
        return "assessment"
    if "probation" in s:
        return "probation"
    return s.strip()

df1_clean["service"] = df1_clean["service"].apply(clean_service_name)
df2_clean["service"] = df2_clean["service"].apply(clean_service_name)
df3_clean["service"] = df3_clean["service"].apply(clean_service_name)

```

```python
df1_clean["service"].unique()
```

### I will only keep "Service Received" Columns

```python
cols_df1 = [c for c in df1_clean.columns if ".1" in c] 
cols_df2 = [c for c in df2_clean.columns if ".1" in c]
cols_df3 = [c for c in df3_clean.columns if ".1" in c]

df1_received = df1_clean[["service"] + cols_df1]
df2_received = df2_clean[["service"] + cols_df2]
df3_received = df3_clean[["service"] + cols_df3]
```

```python
df1_received.head()
```

### Converting strings

```python
for df in [df1_received, df2_received, df3_received]:
    for c in df.columns:
        if c != "service":
            df.loc[:, c] = (
                df[c]
                .astype(str)
                .str.replace(",", "", regex=False)
                .replace("nan", np.nan)
                .astype(float)
            )
```

## Reshape Wide  Long Format

```python
df1_long = df1_received.melt(id_vars="service",
                             var_name="term_year",
                             value_name="count")
df2_long = df2_received.melt(id_vars="service",
                             var_name="term_year",
                             value_name="count")
df3_long = df3_received.melt(id_vars="service",
                             var_name="term_year",
                             value_name="count")
```

```python
df1_long.head()
```

## Convert NaN to Zero After Melt

```python
df1_long["count"] = df1_long["count"].fillna(0)
df2_long["count"] = df2_long["count"].fillna(0)
df3_long["count"] = df3_long["count"].fillna(0)
```

```python
df1_long.head()
```

## Extract Term & Year

```python
def split_term_year(s):
    parts = s.split()
    term = parts[0]
    year = parts[1].replace(".1", "")
    return term, int(year)

df1_long[["term", "year"]] = df1_long["term_year"].apply(
    lambda x: pd.Series(split_term_year(x))
)
df2_long[["term", "year"]] = df2_long["term_year"].apply(
    lambda x: pd.Series(split_term_year(x))
)
df3_long[["term", "year"]] = df3_long["term_year"].apply(
    lambda x: pd.Series(split_term_year(x))
)
```

## Master Dataset

```python
master = pd.concat([df1_long, df2_long, df3_long], ignore_index=True)

master = master[["year", "term", "service", "count"]].dropna()

master.head()
```

## Final Dataset 

```python
master.to_csv("cleaned_student_success_services.csv", index=False)
master.head()
```

## Data Cleaning & Standardization Pipeline Summary

To prepare the Student Success Services dataset for analysis, I built a multi-step cleaning pipeline that standardizes all DataMart files from 20142025 into a single structured dataset. First, I loaded each CSV and removed the top metadata rows that contained descriptive labels instead of data. Then, I extracted the true service names and standardized them (e.g., Credit  Counseling/Advisement Services  counseling). Because DataMart reports each term using four parallel columns (Directed, Service Received, Exempted, Not Received), I selected only the Service Received columns for analysis and converted student counts from string values (e.g., 8,312) into numeric format. I reshaped each dataset from wide format to long format using melt(), resulting in one row per service per term. After converting blank cells to zero and parsing the term and year from column names, I merged all years into one master dataset with consistent columns: year, term, service, count. This cleaned dataset now supports accurate EDA, visualizations, and modeling across the full 20142025 time period.
